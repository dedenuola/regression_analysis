## Import the relevant libraries

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Load the data

```python
data = pd.read_csv(r"C:\Users\user\Downloads\real_estate_price_size.csv")
data.head()
```

## Create the regression

### Declare the dependent and the independent variables

```python
#For this dataset, the dependent variable is the price while the independent variable is the size.
#Hence y is price, while x is size.
y = data['price']
x1 = data['size']
```

### Explore the data

```python
plt.scatter(x1,y, color = 'Red')
plt.title('Scatter Plot of Relationship between Price and Size in Real Estate')
plt.xlabel('Size of an Apartment', fontsize = 10)
plt.ylabel('Cost relative to size', fontsize = 10)
plt.show()
```

### Regression itself

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

### Plot the regression line on the initial scatter

```python
plt.scatter(x1,y)
yhat = x1*223.1787 + 101900
fig = plt.plot(x1,yhat, lw=4, c = 'green', label = 'regression line' )
plt.xlabel('Size', fontsize = 20)
plt.ylabel('Price', fontsize = 20)
plt.show()
```

### Summary of OLS Regression Results
Dependent Variable: Price
Observations: 100 (total number of data points used in the analysis)
Model Fit Statistics:
##### R-squared: 0.745
The result of the R-Squared indicated that approximately 74.5% of the variance in house prices can be explained by the model. This suggests a good fit, as a higher R-squared value usually indicates that the model explains more variance.

##### Adjusted R-squared: 0.742
A value close to 0.742 supports the model's validity.

##### F-statistic: 285.9

Tests the overall significance of the model. A higher F-statistic indicates that the model provides a better fit than a model with no predictors.
Prob (F-statistic): 8.13e-31

##### Size coefficient: 223.1787

For every additional unit (likely square feet) in house size, the price of the house increases by approximately $223.18. This indicates a positive relationship between size and price, meaning larger houses tend to be more expensive.


##### Conclusion:
The regression analysis indicates a strong and statistically significant relationship between the size of the house and its price. The model explains a substantial portion of the variance in house prices, making it a useful tool for predicting prices based on size. Further analysis may be warranted to explore additional predictors or potential model improvements.
